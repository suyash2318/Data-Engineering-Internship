```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv(r"C:\Users\Suyash Maheshwari\Downloads\compressed_data.csv.gz",
                 compression='gzip')

df.head()
```

```python
import sqlite3

conn = sqlite3.connect("superstore.db")

print("Database Connected")
```

```python
df.to_sql("superstore", conn, if_exists="replace", index=False)

print("SQL Table Created")
```

```python
query = """
SELECT *
FROM superstore
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Region='West';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Category='Technology';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Sales > 500;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE "Order Date"
BETWEEN '2016-01-01'
AND '2016-12-31';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Region,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Region;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
AVG(Profit) AS Avg_Profit
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Sub-Category",
SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY "Sub-Category";
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Product Name",
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Customer Name",
SUM(Profit) AS Total_Profit
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Profit DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
strftime('%Y', "Order Date") AS Year,
strftime('%m', "Order Date") AS Month,
SUM(Sales) AS Monthly_Sales
FROM superstore
GROUP BY Year, Month
ORDER BY Year, Month;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT Category,
SUM(Profit) AS Total_Profit
FROM superstore
GROUP BY Category
ORDER BY Total_Profit DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Product Name",
SUM(Profit) AS Total_Profit
FROM superstore
GROUP BY "Product Name"
HAVING Total_Profit < 0
ORDER BY Total_Profit;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Order ID",
"Product ID",
COUNT(*) AS duplicate_count
FROM superstore
GROUP BY "Order ID", "Product ID"
HAVING COUNT(*) > 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Sales IS NULL
OR Profit IS NULL;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM superstore
WHERE Profit < 0;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(DISTINCT "Customer ID")
AS unique_customers
FROM superstore;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT State,
SUM(Profit) AS Total_Profit
FROM superstore
GROUP BY State
ORDER BY Total_Profit DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT "Sub-Category",
SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY "Sub-Category"
ORDER BY Total_Quantity DESC
LIMIT 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(Sales) AS Avg_Sales
FROM superstore;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT MAX(Discount) AS Highest_Discount
FROM superstore;
"""

pd.read_sql(query, conn)
```
